```python
import zipfile

for kind in ['train', 'test']:
    for x in glob.glob(f'./{kind}/*.zip'):
        zipfile.ZipFile(x).extractall(f'./{kind}/')
```

```python
import orjson

table = {
    'train':[],
    'test':[],
}

for kind in ['train', 'test']:
    for x in glob.glob(f'./{kind}/*.json'):
        with open(x, 'rb') as f:
            data = orjson.loads(f.read())
            data = pd.json_normalize(data['Conversation'])
            table[kind].append(data)


```
